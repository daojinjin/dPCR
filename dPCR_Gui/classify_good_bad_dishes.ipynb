```python
import cv2 as cv
# 
import numpy as np
import time
import os
```

```python


good_path = r'F:\sample\good'
```

```python
bad_path = r'F:\sample\bad'
```

```python
samples = []
labels = []
imgs = os.listdir(good_path)
imgs
```

```python
for img in imgs:
    img = cv.imread(good_path +'\\'+ img,2)
    samples.append(img)
    labels.append(1)
imgs = os.listdir(bad_path)
for img in imgs:
    img = cv.imread(bad_path + '\\' +img,2)
    samples.append(img)
    labels.append(0)
print(len(samples))
```

```python
samples_array = np.array(samples).reshape((7811, 28, 28,1))
labels_array = np.array(labels)
samples_array.shape
```

```python
labels_array.shape
```

```python
from sklearn.model_selection import train_test_split
trainX,testX,trainY,testY = train_test_split(samples_array,labels_array,
                                                 test_size=0.3,random_state=420)
```

```python
trainY.shape
```

```python
train_X,validX,train_Y,validY = train_test_split(trainX,trainY,test_size=0.2,random_state=420)
```

```python
train_X.shape
```

```python
from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow.keras import optimizers


model = models.Sequential()


model.add(layers.Conv2D(32,(3,3),activation = 'relu',input_shape = (28,28,1)))
model.add(layers.MaxPooling2D(2,2))
model.add(layers.Conv2D(64,(3,3),activation = 'relu',input_shape = (28,28,1)))
model.add(layers.MaxPooling2D(2,2))
model.add(layers.Conv2D(128,(3,3),activation = 'relu',input_shape = (28,28,1)))
model.add(layers.MaxPooling2D(2,2))


model.add(layers.Flatten())
model.add(layers.Dense(128,activation='relu'))

model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
from tensorflow.keras import optimizers

model.compile(loss='binary_crossentropy',
             optimizer = optimizers.RMSprop(lr=1e-4),
             metrics = ['acc'])
epoch = 50
```

```python
history = model.fit(train_X,train_Y,epochs = epoch,batch_size = 30,validation_data = (validX,validY))
```

```python
model.save('good_bad_dishes.h5')
```
